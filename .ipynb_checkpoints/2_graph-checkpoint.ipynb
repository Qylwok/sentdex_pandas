```python
import pandas as pd

df = pd.read_csv("avocado.csv")
df['Date'] = pd.to_datetime(df["Date"]) # Reformalise les données
albany_df = df[df['region']=="Albany"]
albany_df.set_index("Date", inplace=True)
albany_df["AveragePrice"].plot() # ==> plot un peu dégueu ...
```

```python
albany_df.sort_index(inplace=True) # Car les dates sont dans un ordre chelou °°
albany_df["AveragePrice"].rolling(25).mean().plot()
```

```python
# Pour retirer les warnings quand il commence à y en avoir trop :
albany_df = df.copy()[df["region"] == "Albany"]
# Et on recommence tout (pour l'exemple)
albany_df.set_index("Date", inplace=True)
albany_df.sort_index(inplace=True)
albany_df['price25ma'] = albany_df['AveragePrice'].rolling(25).mean()

# list(set(df['region'].values.tolist()))
df['region'].unique()
```

```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:16]: # 16 seulement parce que sinon on défonce la RAM xddd
    print(region)
    region_df = df.copy()[ df['region'] == region ]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[ [f'{region}_price25ma'] ]
        
    else:
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])
```

```python
# On remédie à ce problème (la jointure marchait pas bien à cause de la colonne type qui contenait 2 données différentes)
import pandas as pd

df = pd.read_csv("avocado.csv")
df = df.copy()[df['type'] == 'organic']
df['Date'] = pd.to_datetime(df["Date"]) # Reformalise les données

df.sort_values(by="Date", ascending=True, inplace=True)

graph_df = pd.DataFrame()

for region in df['region'].unique(): # 16 seulement parce que sinon on défonce la RAM xddd
    print(region)
    region_df = df.copy()[ df['region'] == region ]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[ [f'{region}_price25ma'] ]
        
    else:
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])
    
graph_df.tail()
```

```python
graph_df.dropna().plot(figsize=(15,10), legend=False)
```
